```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("produksi_ayam_jabar.csv")
df.tail(10)
```

```python
df.describe()
```

```python
df.groupby('kode_kabupaten_kota')['nama_kabupaten_kota'].value_counts()
```

```python
df.describe(include='all').T
```

```python
print(df.columns)
```

```python
kota = input(str("Masukan suatu Kota/Kabupaten = ")).lower()
subset = df[df['nama_kabupaten_kota'].str.lower().isin([kota])]
print(subset)
```

```python



produksi_per_tahun = subset.groupby('tahun')['jumlah_produksi'].sum()

plt.plot(produksi_per_tahun.index, produksi_per_tahun.values, color="blue", alpha=0.6, linewidth=2)
total_produksi_kota_kabupaten = np.sum(produksi_per_tahun)
rata2_produksi_kota_kabupaten = np.mean(produksi_per_tahun)

plt.title('Jumlah Total Produksi per Tahun di ' + kota)
plt.xlabel('Tahun')
plt.ylabel('Jumlah Produksi')
plt.grid(axis='y') # Menambah garis di bagian y-axis
plt.show()


print("Rata-rata produksi di " + kota + "=", rata2_produksi_kota_kabupaten)
print("Total produksi di " + kota + "=", total_produksi_kota_kabupaten)
```

```python
produksi_jawa_barat = df.groupby('nama_kabupaten_kota')['jumlah_produksi'].sum()

plt.figure(figsize=(12, 8))  # Menentukan ukuran plot
plt.bar(produksi_jawa_barat.index, produksi_jawa_barat.values, color='cyan')
total_produksi_jawa_barat = np.sum(produksi_jawa_barat)
rata2_produksi_jawa_barat = np.mean(produksi_jawa_barat)

plt.title('Jumlah Total Produksi di Jawa Barat dari tahun 2014 sampai 2022')
plt.xlabel('Nama Kabupaten/Kota')
plt.ylabel('Jumlah Produksi')
plt.xticks(rotation=80)
plt.grid(axis='y')

plt.show()

print("Rata-rata produksi di Jawa Barat:", rata2_produksi_jawa_barat)
print("Total produksi di Jawa Barat:", total_produksi_jawa_barat)
```

```python
tahun_berapa = int(input('Masukan data tahun berapa(2014 - 2022) = '))
data_pertahun = df[df['tahun'] == tahun_berapa]

produksi_pertahun = data_pertahun.groupby('nama_kabupaten_kota')['jumlah_produksi'].sum()

plt.figure(figsize=(12, 6))  # angka pertama (10) menentukan lebar figur (figure width) dalam satuan inch, sedangkan angka kedua (6) menentukan tinggi figur (figure height) dalam satuan inch.
sns.scatterplot( color='red',x=produksi_pertahun.index, y=produksi_pertahun.values)
total_produksi_jawa_barat_pertahun = np.sum(produksi_pertahun)
rata2_produksi_jawa_barat_pertahun = np.mean(produksi_pertahun)

plt.title('Jumlah Total Produksi di seluruh Jawa Barat Tahun %d' % tahun_berapa)
plt.xlabel('Nama Kabupaten/Kota')
plt.ylabel('Jumlah Produksi')
plt.xticks(rotation=80)
plt.grid(axis='y')

plt.show()

print("Rata-rata di Jawa Barat Tahun %d:" % tahun_berapa, rata2_produksi_jawa_barat_pertahun)
print("Total produksi di Jawa Barat Tahun %d:" % tahun_berapa, total_produksi_jawa_barat_pertahun)
```

```python
total_produksi_per_tahun = df.groupby('tahun')['jumlah_produksi'].sum().reset_index()

plt.figure(figsize=(12,8))
sns.barplot(data=total_produksi_per_tahun, x='tahun', y='jumlah_produksi', hue='tahun', palette='viridis')
plt.title('Total Produksi per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Total Produksi')
plt.grid(axis='y')
plt.show()
```
